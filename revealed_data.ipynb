```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', lambda x: '%.2f' % x)


DATA_PATH = '/Users/kaiqu/kaggle-datasets/example_test_files/revealed_targets.csv'

# Load data from CSV
df = pd.read_csv(DATA_PATH)
```

```python
df.info()
```

```python
df[df['stock_id'] == 1].describe()
```

```python
df[df['stock_id'] == 1].isnull().sum()
```

```python
df_stock_1 = df[df['stock_id'] == 1.0]
# Plotting the distribution of the Revealed Target for stock_id = 1
plt.figure(figsize=(10, 6))
sns.histplot(df_stock_1['revealed_target'].dropna(), bins=50, kde=True)
plt.title('Distribution of Revealed Target for Stock ID = 1')
plt.xlabel('Revealed Target')
plt.ylabel('Frequency')
plt.show()

# Boxplot for the Revealed Target for stock_id = 1
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_stock_1['revealed_target'])
plt.title('Boxplot of Revealed Target for Stock ID = 1')
plt.xlabel('Revealed Target')
plt.show()
```

```python
df_stock_1.to_csv("revealed_data_stock_1.csv")
```
